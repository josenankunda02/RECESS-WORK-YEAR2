```python
# reading the dataset
# encoding has to be changed since the file contains non utf8 characters
profile=pd.read_csv('mps.csv', encoding='latin-1')
profile
```

```python
#dataset before cleaning
```

```python
# Determining dataframe shape
profile.shape
```

```python
# Detailed information about dataframe
profile.info()
```

```python
# checking for nulls
profile.isnull().sum().sort_values(ascending=False)
```

```python
#Data cleaning
#Empty columns
# Drop the 11 empty columns
# Assuming you have a DataFrame named 'df'
threshold = len(profile) - 300

profile.dropna(axis=1, thresh=threshold, inplace=True)

#drop the first column because its unnecessarily its repeatitive

profile = profile.drop('No.', axis=1)
profile
```

```python
#Dataset after cleaning
# new shape
#columns reduced from 87 to 34
profile.shape
```

```python
#New columns after removing the null columns
profile.columns
```

```python
# handling Wrong format for date of birth
#here is the correct dataset

profile['Date of Birth'] = pd.to_datetime(profile['Date of Birth'], errors='coerce')


profile['Date of Birth'] = profile['Date of Birth'].dt.strftime('%m/%d/%Y')
profile
```

```python
#PROFILE VISUALIZATION
#Piechart
#Piechart
data = profile['Political Party'].value_counts()
labels = data.index.tolist()
plt.pie(data, labels=labels, autopct='%5.1f%%')
plt.title('Pie Chart For Political Parties')
plt.show()
```

```python
#Histogram
# Group the data by 'Marital Status'
grouped_data = profile.groupby('Marital status')

# List of unique marital statuses
marital_statuses = profile['Marital status'].unique()

# List of colors for each marital status
colors = ['green', 'blue', 'red']

# Plot histograms for each marital status
for status, color in zip(marital_statuses, colors):
    data = grouped_data.get_group(status)
    plt.hist(data['Marital status'], bins=10, color=color, edgecolor='black', label=status)

plt.xlabel('Marital status')
plt.ylabel('Frequency')
plt.title('Histogram by Marital Status')
plt.legend()
plt.show()
```
