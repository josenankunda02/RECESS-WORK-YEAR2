```python
 import pandas as pd
import numpy as np
#1.Write a Pandas program to compare the elements of two Pandas series.
# Sample Series [4,65,436,3,9], [7,0,3,897,9]
series1 = pd.Series([4, 65, 436, 3, 9])
series2 = pd.Series([7, 0, 3, 897, 9])

comparison = series1 == series2
print(comparison)
```

```python
# 2.Write a Pandas program to add, subtract, multiply and divide two Pandas series. 
# Sample Series: [2,4,6,8,14], [1,3,5,7,9]
series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])

addition = series1 + series2
subtraction = series1 - series2
multiplication = series1 * series2
division = series1 / series2

print("Addition:")
print(addition)
print("\nSubtraction:")
print(subtraction)
print("\nMultiplication:")
print(multiplication)
print("\nDivision:")
print(division)
```

```python
# 6.Write a Pandas program to select the rows where the ‘Attendance’ is missing.
missing_attendance_rows = data[data['Attendance'].isnull()]

print(missing_attendance_rows)
```

```python
# 7.Write a Pandas program to select the rows where the ‘HomeTeamGoals’ are between 3 and 6 inclusive.
selected_rows = data[(data['HomeTeamGoals'] >= 3) & (data['HomeTeamGoals'] <= 6)]

print(selected_rows)
```

```python
# 9.Write a Pandas program to sort the DataFrame first by ‘HomeTeam’ in ascending order, then by ‘HomeTeamScores’ in descending order.
sorted_df = data.sort_values(['HomeTeam', 'HomeTeamGoals'], ascending=[True, False])

print(sorted_df)
```

```python
# 12.Write a Pandas program to add 2 rows to your DataFrame.


new_rows = [{'HomeTeam': 'TeamA', 'AwayTeam': 'TeamB', 'HomeTeamGoals': 2, 'AwayTeamGoals': 1},
            {'HomeTeam': 'TeamX', 'AwayTeam': 'TeamY', 'HomeTeamGoals': 3, 'AwayTeamGoals': 2}]

data = pd.concat([data, pd.DataFrame(new_rows)], ignore_index=True)

pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
data
```
