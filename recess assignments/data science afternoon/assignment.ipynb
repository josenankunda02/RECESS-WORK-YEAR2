# Assignment three

```python
# reading the dataset
df_uce=pd.read_csv('UceResultsBySchool2011-2016.csv')
# Set the option to display all rows and columns
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)
df_uce
```

```python
# Detailed dataframe info
df_uce.info()
```

```python
# checking for nulls
# none dropped
df_uce.isnull().sum().sort_values(ascending=False)
```

#Removing empty columns

```python
# Drop the 11 empty columns
df_uce.dropna(axis=1, how='all',inplace=True)
df_uce
```

Filling in empty cells

```python
# Finding mean for numerical columns with missing values less than half the total values

# Calculate total number of values in each column
total_values = df_uce.shape[0]

# Determine threshold for missing values
threshold = total_values / 4

# Filter numerical columns based on missing values threshold
numerical_columns = df_uce.select_dtypes(include=np.number).columns
filtered_columns = []
for column in numerical_columns:
    missing_values = df_uce[column].isnull().sum()
    if missing_values < threshold:
        filtered_columns.append(column)

# Calculate mean for filtered numerical columns
column_means = df_uce[filtered_columns].mean().astype(int)

print(column_means)
```

```python
# Filling empty cells of numerical columns for the above
df_uce[filtered_columns] = df_uce[filtered_columns].fillna(column_means)
df_uce   #new dataframe
```

```python
# new dataframe info showing less nulls
df_uce.info()
```

```python
# check nulls to confirm some numerical columns are filled 
df_uce.isnull().sum().sort_values(ascending=True)
```

```python
# Finding mode for numerical columns with missing values between a quarter and three quarters of the total values

# Calculate total number of values in each column
total_values = df_uce.shape[0]

# Determine threshold for missing values
lower_threshold = total_values / 4
upper_threshold = 3 * total_values / 4

# Filter numerical columns based on missing values threshold
numerical_columns = df_uce.select_dtypes(include=np.number).columns
filtered_columns = []
for column in numerical_columns:
    missing_values = df_uce[column].isnull().sum()
    if lower_threshold <= missing_values <= upper_threshold:
        filtered_columns.append(column)

# Calculate modefor filtered numerical columns
column_modes = df_uce[filtered_columns].mode().astype(int)

column_modes
```

```python
# Filling empty cells of numerical columns for the above
df_uce[filtered_columns] = df_uce[filtered_columns].fillna(column_modes)
df_uce   #new dataframe
```

```python
# check nulls to confirm some numerical columns are filled 
df_uce.isnull().sum().sort_values(ascending=True)
```

```python
# Replace the rest of numerical columns with empty cells the mode
# Select only numerical columns
numerical_columns = df_uce.select_dtypes(include=np.number).columns

# Calculate median for numerical columns
column_medians = df_uce[numerical_columns].median().astype(int)
column_medians

```

```python
# Fill missing values in numerical columns
df_uce[numerical_columns] = df_uce[numerical_columns].fillna(column_medians)  # Replace the rest of empty cells with mode

df_uce
```

```python
df_uce.info()     # new dataframe info
```

```python
# check nulls to confirm all numerical columns are filled now
df_uce.isnull().sum().sort_values(ascending=True)
```

```python
#  Replacing empty cell in SCHOOL column


# Fill missing string values in a column with the previous non-null value
df_uce['SCHOOL'] = df_uce['SCHOOL'].fillna(method='ffill')

# Check the updated DataFrame for missing values
print(df_uce.isnull().sum())
```
