```python
#WEB SCRAPPING
from bs4 import BeautifulSoup
import requests

#EXTRACTING ALL BIRD SPECIES FROM THE WEBSITE
def scrape_bird_species(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the table containing the bird species
    species_table = soup.find('table', class_='results')
    
    # Initialize an empty list to store the bird species
    bird_species_list = []
    
    # Iterate through each row in the table (skipping the header row)
    for row in species_table.find_all('tr')[1:]:
        # Extract the species name from the first cell in each row
        species_name = row.find_all('td')[0].text.strip()
        
        # Append the species name to the list
        bird_species_list.append(species_name)
    
    # Store the list of bird species inside a dictionary
    bird_species_dict = {"bird_species": bird_species_list}
    
    return bird_species_dict

# URL for bird species data extraction
species_url = "https://xeno-canto.org/collection/species/all"

# Scrape bird species and organize them in a single list within a dictionary
bird_species_data = scrape_bird_species(species_url)

# Print the dictionary of bird species
print(bird_species_data)
```

```python
#List of Birds without duplicates
def scrape_bird_species(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the table containing the bird species
    species_table = soup.find('table', class_='results')
    
    # Initialize an empty list to store the bird species
    bird_species_list = []
    
    # Iterate through each row in the table (skipping the header row)
    for row in species_table.find_all('tr')[1:]:
        # Extract the species name from the first cell in each row
        species_name = row.find_all('td')[0].text.strip()
        
        # Append the species name to the list
        bird_species_list.append(species_name)
    
    # Remove duplicates by converting the list to a set and back to a list
    bird_species_list = list(set(bird_species_list))
    
    # Store the list of bird species inside a dictionary
    bird_species_dict = {"bird_species": bird_species_list}
    
    return bird_species_dict

# URL for bird species data extraction
species_url = "https://xeno-canto.org/collection/species/all"

# Scrape bird species and organize them in a single list within a dictionary
bird_species_data = scrape_bird_species(species_url)

# Print the dictionary of bird species (with no duplicates in the list)
print(bird_species_data)
```

```python
# Formattimg list of Bird species without duplicates into a CSV file

import csv

def scrape_bird_species(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the table containing the bird species
    species_table = soup.find('table', class_='results')
    
    # Initialize an empty list to store the bird species
    bird_species_list = []
    
    # Iterate through each row in the table (skipping the header row)
    for row in species_table.find_all('tr')[1:]:
        # Extract the species name from the first cell in each row
        species_name = row.find_all('td')[0].text.strip()
        
        # Append the species name to the list
        bird_species_list.append(species_name)
    
    # Remove duplicates by converting the list to a set and back to a list
    bird_species_list = list(set(bird_species_list))
    
    return bird_species_list

# URL for bird species data extraction
species_url = "https://xeno-canto.org/collection/species/all"

# Scrape bird species and get the list with no duplicates
bird_species_data = scrape_bird_species(species_url)

# Write the list to a CSV file
csv_file = "bird_species.csv"
with open(csv_file, "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Bird Species"])
    writer.writerows(zip(bird_species_data))

print(f"Data has been written to {csv_file}.")
```
